# **4. Background and rationale**

## **4.1 Health issue or knowledge gap**

Chronic respiratory disease includes a range of conditions that affect multiple components of respiratory system, including the airways, parenchyma, and pulmonary vasculature. There are two types of chronic respiratory disease, communicable and noncommunicable. Communicable Chronic respiratory disease is long-term respiratory condition caused by infectious agents, while noncommunicable chronic respiratory disease are long-term respiratory conditions caused from lifestyle factors, chronic inflammation and environmental exposure. Some of the noncommunicable respiratory disease includes asthma, chronic obstructive pulmonary disease (COPD), lung cancer and others. Asthma is a major noncommunicable respiratory disease, and it has affected an estimated 262 million people in 2019, making it increasingly common these days, and it has caused 455,000 deaths according to WHO. It is also often under-diagnosed and under-treated in rural areas (with low income). To diagnose for the functioning of lungs especially when monitoring Chronic obstructive pulmonary disease, we can use spirometers to determine the ventilatory capacity of the lungs. Healthy lung size and function is dependent on four key variables, which are age, sex, height, and race, and the lung function values from the individual will be compared to normal patients with similar demographics, and if the value is below the normal range, it is considered abnormal. Some of the measurement used to interpret spirometry includes forced vital capacity (VC), forced expiratory volume in the first second (FEV 1), FEV1/FVC ratio and peak expiratory flow (PEF). In addition, elevated levels of C-reactive protein (CRP), a marker of low-grade systemic inflammation is also found associated in asthmatic individuals.


## **4.2 Public health and biomedical importance**

This research question is important in both public health and biomedical science because it aims to understand how the different risk factors influence respiratory health and inflammation among adults with and without asthma. By identifying how factors like smoking and BMI affect the condition of the lung function in individuals with and without asthma, we can contribute to better risk stratification and promote tailored strategies to reduce the risk of chronic respiratory disease.

## **4.3 Use of NHANES and extension of existing work**

This project utilizes the data from National Health and Nutrition Examination Survey (NHANES) 2007-2018 which provides us information on spirometry results, CRP levels, BMI, environmental exposure, smoking behaviour socio-economic status, gender. While previous studies have shown individual risk factors affecting lung function, this project extends existing work by analysing multiple variables simultaneously and comparing these associations between individuals with and without asthma allowing for more insights. The use of multiple biomarkers such as FEV1, FEV1/FVC ratio, peak expiratory flow and CRP levels also provides a multidimensional view of respiratory health. We hope that this analysis will clarify how lifestyle and socioeconomic affect lung function, helping to better determine better preventative actions towards respiratory diseases, particularly asthma  

# **5. Research question and Objectives**

## **5.1 Research question**

- Using NHANES 2007-2018, among U.S. adults aged 18+, how do smoking status and BMI predict reduced FEV1/FVC ratio in individuals with and without prior asthma diagnosis?

## **5.2 Justification and novelty**

Reduced lung function leading to respiratory disease can be measured by the FEV1/FVC ratio and it is influenced by various lifestyle, environmental and demographic factors. Smoking is known for being a risk factor for respiratory disease, but different smoking status (e.g. occasional smoker, regular smoker, quite smoking less than or more than 1 year ago) has different impacts on the lungs. BMI has also been associated with impaired respiratory function and systemic inflammation, which can further affect respiratory health in both non-asthmatic and asthmatic individuals. There has been research on the independent effects of smoking and BMI on lung health, but not a lot have explored how these factors jointly influence FEV1/FVC ratio in subgroups defined by asthma status.

## **5.3 Contribution to existing literature bold text**

Previous studies have shown that in non-obese individuals, BMI does not significantly affect lung function, whereas in obese individuals, BMI does affect lung function. It is also shown that non-asthmatic smokers face greater decline in FEV1 than non-asthmatic non-smokers, and asthmatic smokers face greater FEV1 decline than non-asthmatic smokers. Furthermore, it has been observed that increasing BMI is associated to lower FEV1/FVC ratios, suggesting that obesity is an independent factor that affects impairment of lung function. It however remains unclear if other risk factors such as smoking could influence this relationship, and whether asthmatics and non-asthmatics respond the same to changes in these independent factors. Through this research project, we are extending existing research by:

1.   Comparing the difference in the degree of FEV1/FVC ratio being affected by BMI and smoking in asthmatics and non-asthmatics
2.   Investigate the effects of Smoking, BMI and asthma status on FEV1/FVC ratio

## **5.4 SMART Objectives**

1. This study aims to investigate how smoking status and BMI predict reduced FEV1/FVC ratio in U.S. adults aged 18 and above — using measurable variables from NHANES 2007–2018 — with a feasible, time-bound analysis that addresses a relevant public health concern.

2. By analysing 12 years of NHANES data, this study will quantify how smoking and BMI contribute to impaired lung function in adults with and without asthma, providing attainable and time-bound evidence for targeted health strategies.

3. The research will use national survey data from 2007–2018 to specifically assess the predictive impact of smoking status and BMI on reduced FEV1/FVC ratio, producing relevant and measurable findings for respiratory health policy.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Step 0: Define years and NHANES cycles
# -----------------------------
years = ["2007", "2009", "2011", "2013", "2015", "2017"]
cycles = ["E", "F", "G", "H", "I", "J"]

demo_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/DEMO_{c}.xpt"
             for y, c in zip(years, cycles)]
df_demo = pd.concat([pd.read_sas(url) for url in demo_urls], ignore_index=True)

df_demo = df_demo[["SEQN", "RIAGENDR", "RIDAGEYR", "RIDRETH1", "INDFMPIR"]]

bmi_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/BMX_{c}.xpt"
            for y, c in zip(years, cycles)]
df_bmi = pd.concat([pd.read_sas(url) for url in bmi_urls], ignore_index=True)

# Keep only necessary columns
df_bmi = df_bmi[["SEQN", "BMXBMI"]]

asthma_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/MCQ_{c}.xpt"
               for y, c in zip(years, cycles)]
df_asthma = pd.concat([pd.read_sas(url) for url in asthma_urls], ignore_index=True)

df_asthma = df_asthma[["SEQN", "MCQ010"]]  # MCQ010: "Ever told you had asthma?"

smq_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/SMQ_{c}.xpt"
            for y, c in zip(years, cycles)]
df_smq = pd.concat([pd.read_sas(url) for url in smq_urls], ignore_index=True)

df_smq = df_smq[["SEQN", "SMQ020", "SMQ040"]]  # Smoking status variables

# Spirometry (FEV1 & FVC)
# -----------------------------
spirometry_years   = ["2007", "2009", "2011"]
spirometry_cycles  = ["E", "F", "G"]          # Only 2007–2012 have spirometry
spx_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/SPX_{c}.xpt"
            for y, c in zip(spirometry_years, spirometry_cycles)]

df_spiro = pd.concat([pd.read_sas(url) for url in spx_urls])

# keep baseline FEV1 & FVC
df_spiro = df_spiro[["SEQN", "SPXNFEV1", "SPXNFVC"]]
df_spiro = df_spiro.rename(columns={"SPXNFEV1": "FEV1", "SPXNFVC": "FVC"})

paq_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/PAQ_{c}.xpt"
            for y, c in zip(years, cycles)]
df_paq = pd.concat([pd.read_sas(url) for url in paq_urls], ignore_index=True)

df_paq = df_paq[["SEQN", "PAD660", "PAD675", "PAD680"]]



# Merge all datasets sequentially on SEQN, keeping all demographics records (left join)
df = (df_demo
      .merge(df_bmi, on="SEQN", how="left")
      .merge(df_asthma, on="SEQN", how="left")
      .merge(df_smq, on="SEQN", how="left")
      .merge(df_spiro, on="SEQN", how="left")
      .merge(df_paq, on="SEQN", how="left"))


#Step 8:  categorize smokers
# -----------------------------

#SMQ020: Ever smoked 100 cigarettes in life?

#1 = Yes (Smoker)
#2 = No (Non-smoker)

#SMQ040: Do you now smoke cigarettes?
#1 = Every day
#2 = Some days
#3 = Not at all


def classify_smoking(row):
    if pd.isna(row["SMQ020"]):
        return np.nan
    if row["SMQ020"] == 2:
        return "Never smoked"
    elif row["SMQ020"] == 1:
        if pd.isna(row["SMQ040"]):
            return np.nan
        if row["SMQ040"] == 1:
            return "Regular smoker"
        elif row["SMQ040"] == 2:
            return "Occasional smoker"
        elif row["SMQ040"] == 3:
            return "Reformed non-smoker"
    return np.nan


#First drop

df["Smoking"] = df.apply(classify_smoking, axis=1)

df.drop(columns=["SMQ020", "SMQ040"], inplace=True)

print (df["Smoking"].value_counts())
#

import numpy as np

# Step 1: Clean values
df["PAD660"] = df["PAD660"].replace({777: np.nan, 999: np.nan})
df["PAD675"] = df["PAD675"].replace({777: np.nan, 999: np.nan})
df["PAD680"] = df["PAD680"].replace({77: np.nan, 99: np.nan})

# Step 2: Categorize activity level
def classify_activity(row):
    has_vigorous = not pd.isna(row["PAD660"]) and row["PAD660"] > 0
    has_moderate = not pd.isna(row["PAD675"]) and row["PAD675"] > 0

    if has_vigorous and has_moderate:
        return "Moderate and vigorous activity"
    elif has_vigorous:
        return "Vigorous activity only"
    elif has_moderate:
        return "Moderate activity only"
    else:
        return "No moderate or vigorous activity"

# Step 3: Apply classification
df["Exercise"] = df.apply(classify_activity, axis=1)

# Optional: Drop raw activity columns
# df.drop(columns=["PAD660", "PAD675", "PAD680"], inplace=True)

# Step 4: Check result
print(df["Exercise"].value_counts(dropna=False))



def categorize_bmi(bmi):
    if pd.isna(bmi):
        return np.nan
    elif bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Normal"
    elif 25 <= bmi < 30:
        return "Overweight"
    else:
        return "Obese"

# Apply function to create new column
df["BMI_category"] = df["BMXBMI"].apply(categorize_bmi)

# Optional: drop original BMI if not needed
# df.drop(columns=["BMXBMI"], inplace=True)

# Check value counts
print(df["BMI_category"].value_counts())


def categorize_income(pir):
    if pd.isna(pir):
        return np.nan
    elif pir <= 1.0:
        return "At or below poverty"
    elif pir <= 2.0:
        return "Low income"
    elif pir <= 3.5:
        return "Middle income"
    else:
        return "High income"

# Apply function to create new SES category column
df["SES_Category"] = df["INDFMPIR"].apply(categorize_income)

# Optional: drop original PIR variable if not needed
# df.drop(columns=["INDFMPIR"], inplace=True)

# Check value counts
print(df["SES_Category"].value_counts(dropna=False))


# Define asthma classification function
def classify_asthma(row):
    if pd.isna(row["MCQ010"]):
        return np.nan
    elif row["MCQ010"] == 1:
        return "Asthma"
    elif row["MCQ010"] == 2:
        return "No asthma"
    elif row["MCQ010"] in [7, 9]:
        return np.nan
    return np.nan

# Apply classification to df_asthma
df_asthma["Asthma_Status"] = df_asthma.apply(classify_asthma, axis=1)

# Optionally drop MCQ010
df_asthma.drop(columns=["MCQ010"], inplace=True)

# View counts
print(df_asthma["Asthma_Status"].value_counts(dropna=False))


df_spiro["FEV1_FVC_ratio"] = df_spiro["FEV1"] / df_spiro["FVC"]
def classify_spirometry(row):
    if pd.isna(row["FEV1"]) or pd.isna(row["FVC"]):
        return np.nan
    if row["FEV1_FVC_ratio"] < 0.70:
        return "Obstructive"
    else:
        return "Normal/Non-obstructive"
df_spiro["Spirometry_Category"] = df_spiro.apply(classify_spirometry, axis=1)

# Optionally, drop raw ratio column
df_spiro.drop(columns=["FEV1_FVC_ratio"], inplace=True)

print(df_spiro["Spirometry_Category"].value_counts())



df.drop (columns=["SEQN"], inplace=True)
#Restrict age to >18
df = df[(df["RIDAGEYR"] >= 18)]

#For all columns replace "Unknown" with np.nan
df.replace("Unknown", np.nan, inplace=True)
#Define the categorical columns
cat_cols = ["RIAGENDR", "Smoking", "Exercise", "BMI_category", "SES_Category", "Asthma_Status", "Spirometry_Category", "CRP_Category" ]
#Define the numerical columns
num_cols = ["RIDAGEYR", "BMXBMI"]

print ("\nSize of dataset: ", "Columns: ", len(df.columns), "Rows: ", len(df), "\n")

# Summarize only the numerical columns
display (df[num_cols].describe())

#Explore missing data
missing_data = df.isnull().sum()
print ("\nMissing data")
print(missing_data)
print(df.columns.tolist())


```

```python
#Effect of both smoking and BMI on asthmatics and non-asthmatics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import itertools
import seaborn as sns

# 1. Prepare Data

# Map MCQ010 to Asthma_Status if not already done
if 'Asthma_Status' not in df.columns:
    df['Asthma_Status'] = df['MCQ010'].map({1: 'Asthma', 2: 'No asthma'})

# Create FEV1/FVC ratio if not already created
if 'FEV1_FVC_ratio' not in df.columns:
    df['FEV1_FVC_ratio'] = df['FEV1'] / df['FVC']

# Drop rows with missing values in relevant columns
df_interact = df.dropna(subset=['FEV1_FVC_ratio', 'Smoking', 'BMXBMI', 'Asthma_Status'])

# 2. Fit interaction regression model
model_interact = smf.ols('FEV1_FVC_ratio ~ C(Smoking) * BMXBMI * C(Asthma_Status)', data=df_interact).fit()

print(model_interact.summary())

# 3. Prepare prediction grid for plotting

asthma_categories = df_interact['Asthma_Status'].unique()
smoking_categories = df_interact['Smoking'].unique()

# Generate BMI values over observed range
bmi_min, bmi_max = df_interact['BMXBMI'].min(), df_interact['BMXBMI'].max()
bmi_vals = np.linspace(bmi_min, bmi_max, 100)

# Create DataFrame with all combinations of BMI, Smoking, Asthma_Status for prediction
pred_grid = pd.DataFrame(list(itertools.product(bmi_vals, smoking_categories, asthma_categories)),
                         columns=['BMXBMI', 'Smoking', 'Asthma_Status'])

# Predict FEV1/FVC ratio using the model
pred_grid['Predicted_FEV1_FVC'] = model_interact.predict(pred_grid)

# 4. Plot predicted FEV1/FVC by BMI and Smoking status for Asthma and No asthma

color_cycle = sns.color_palette("tab10")
marker_cycle = itertools.cycle(['o', 's', '^', 'D', 'v', '*'])

import math
num_asthma = len(asthma_categories)
cols = 2
rows = math.ceil(num_asthma / cols)

plt.figure(figsize=(cols * 7, rows * 5))

for i, asthma in enumerate(sorted(asthma_categories)):
    plt.subplot(rows, cols, i+1)
    for j, smoking in enumerate(smoking_categories):
        subset = pred_grid[(pred_grid['Asthma_Status'] == asthma) & (pred_grid['Smoking'] == smoking)]
        color = color_cycle[j % len(color_cycle)]
        marker = next(marker_cycle)
        plt.plot(subset['BMXBMI'], subset['Predicted_FEV1_FVC'],
                 label=smoking, color=color, marker=marker, markevery=10)

    plt.title(f'Predicted FEV1/FVC by BMI and Smoking\n(Asthma Status: {asthma})')
    plt.xlabel('BMI')
    plt.ylabel('Predicted FEV1/FVC Ratio')
    plt.ylim([0.5, 1])
    plt.legend(title='Smoking Status')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Association of income category on asthma presence
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import chi2_contingency
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Categorize income ratio into brackets
bins = [0, 1, 2, 3, 5, 10, 100]
labels = ['<1 (Below Poverty)', '1-2', '2-3', '3-5', '5-10', '10+']
df['Income_Category'] = pd.cut(df['INDFMPIR'], bins=bins, labels=labels)

# Clean data for chi-square
df_cat = df[['Income_Category', 'Asthma_Status']].dropna()

# Create contingency table
contingency_table = pd.crosstab(df_cat['Income_Category'], df_cat['Asthma_Status'])
print("Contingency Table:")
print(contingency_table)

# Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"\nChi-square test statistic: {chi2:.3f}")
print(f"P-value: {p:.4f}")
if p < 0.05:
    print("=> There is a significant association between income category and asthma presence.")
else:
    print("=> No significant association between income category and asthma presence.")

# Prepare data for logistic regression
df['Asthma_Binary'] = df['Asthma_Status'].map({'Asthma':1, 'No asthma':0})
df_logit = df.dropna(subset=['INDFMPIR', 'Asthma_Binary'])

# Logistic regression
model = smf.logit('Asthma_Binary ~ INDFMPIR', data=df_logit).fit()
print(model.summary())

# ▶️ Stacked bar chart: Asthma presence by income category
contingency_table.plot(kind='bar', stacked=True, figsize=(10,6), color=['skyblue', 'salmon'])
plt.title('Asthma Presence by Income Category')
plt.xlabel('Income Category (Family Income to Poverty Ratio)')
plt.ylabel('Count')
plt.legend(title='Asthma Status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import io
import seaborn as sns

# -----------------------------
# Step 0: Define years and NHANES cycles
# -----------------------------
years = ["2007", "2009", "2011", "2013", "2015", "2017"]
cycles = ["E", "F", "G", "H", "I", "J"]

demo_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/DEMO_{c}.xpt"
             for y, c in zip(years, cycles)]
df_demo = pd.concat([pd.read_sas(url) for url in demo_urls], ignore_index=True)
df_demo = df_demo[["SEQN", "RIAGENDR", "RIDAGEYR", "RIDRETH1", "INDFMPIR"]]

bmi_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/BMX_{c}.xpt"
            for y, c in zip(years, cycles)]
df_bmi = pd.concat([pd.read_sas(url) for url in bmi_urls], ignore_index=True)
df_bmi = df_bmi[["SEQN", "BMXBMI"]]

asthma_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/MCQ_{c}.xpt"
               for y, c in zip(years, cycles)]
df_asthma = pd.concat([pd.read_sas(url) for url in asthma_urls], ignore_index=True)
df_asthma = df_asthma[["SEQN", "MCQ010"]]

smq_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/SMQ_{c}.xpt"
            for y, c in zip(years, cycles)]
df_smq = pd.concat([pd.read_sas(url) for url in smq_urls], ignore_index=True)
df_smq = df_smq[["SEQN", "SMQ020", "SMQ040"]]

paq_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/PAQ_{c}.xpt"
            for y, c in zip(years, cycles)]

df_paq = pd.concat([pd.read_sas(url) for url in paq_urls], ignore_index=True)

# Keep the three activity-count variables you use later
df_paq = df_paq[["SEQN", "PAD660", "PAD675", "PAD680"]]


# CRP / hs-CRP (final version)
def get_crp_url_and_var(year, letter):
    if letter in ["E", "F"]:
        return (
            f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/CRP_{letter}.xpt",
            "LBXCRP"
        )
    elif letter in ["I", "J"]:
        return (
            f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/HSCRP_{letter}.xpt",
            "LBXHSCRP"
        )
    else:
        return None

crp_frames = []
for y, c in zip(years, cycles):
    crp_info = get_crp_url_and_var(y, c)
    if crp_info is None:
        print(f"No CRP measured in cycle {c} ({y}) – skipping")
        continue

    url, varname = crp_info
    try:
        buf = io.BytesIO(requests.get(url, timeout=60).content)
        df_tmp = pd.read_sas(buf, format="xport")[["SEQN", varname]]
        df_tmp = df_tmp.rename(columns={varname: "LBXCRP"})  # unify column name
        crp_frames.append(df_tmp)
        print(f"Loaded {url.split('/')[-1]}")
    except Exception as e:
        print(f"Could not load {url.split('/')[-1]} – {e}")

df_crp = pd.concat(crp_frames, ignore_index=True)

df = (df_demo
      .merge(df_bmi,    on="SEQN", how="left")
      .merge(df_asthma, on="SEQN", how="left")
      .merge(df_smq,    on="SEQN", how="left")
      .merge(df_paq,    on="SEQN", how="left")
      .merge(df_crp,    on="SEQN", how="left"))

# Step 8: categorize smokers
def classify_smoking(row):
    if pd.isna(row["SMQ020"]):
        return np.nan
    if row["SMQ020"] == 2:
        return "Never smoked"
    elif row["SMQ020"] == 1:
        if pd.isna(row["SMQ040"]):
            return np.nan
        if row["SMQ040"] == 1:
            return "Regular smoker"
        elif row["SMQ040"] == 2:
            return "Occasional smoker"
        elif row["SMQ040"] == 3:
            return "Reformed non-smoker"
    return np.nan

df["Smoking"] = df.apply(classify_smoking, axis=1)
df.drop(columns=["SMQ020", "SMQ040"], inplace=True)
print(df["Smoking"].value_counts())

# Clean exercise variables
df["PAD660"] = df["PAD660"].replace({777: np.nan, 999: np.nan})
df["PAD675"] = df["PAD675"].replace({777: np.nan, 999: np.nan})
df["PAD680"] = df["PAD680"].replace({77: np.nan, 99: np.nan})

def classify_activity(row):
    has_vigorous = not pd.isna(row["PAD660"]) and row["PAD660"] > 0
    has_moderate = not pd.isna(row["PAD675"]) and row["PAD675"] > 0

    if has_vigorous and has_moderate:
        return "Moderate and vigorous activity"
    elif has_vigorous:
        return "Vigorous activity only"
    elif has_moderate:
        return "Moderate activity only"
    else:
        return "No moderate or vigorous activity"

df["Exercise"] = df.apply(classify_activity, axis=1)

# Categorize BMI
def categorize_bmi(bmi):
    if pd.isna(bmi):
        return np.nan
    elif bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Normal"
    elif 25 <= bmi < 30:
        return "Overweight"
    else:
        return "Obese"

df["BMI_category"] = df["BMXBMI"].apply(categorize_bmi)

# Categorize income (SES)
def categorize_income(pir):
    if pd.isna(pir):
        return np.nan
    elif pir <= 1.0:
        return "At or below poverty"
    elif pir <= 2.0:
        return "Low income"
    elif pir <= 3.5:
        return "Middle income"
    else:
        return "High income"

df["SES_Category"] = df["INDFMPIR"].apply(categorize_income)

# Asthma classification
def classify_asthma(row):
    if pd.isna(row["MCQ010"]):
        return np.nan
    elif row["MCQ010"] == 1:
        return "Asthma"
    elif row["MCQ010"] == 2:
        return "No asthma"
    elif row["MCQ010"] in [7, 9]:
        return np.nan
    return np.nan

df["Asthma_Status"] = df.apply(classify_asthma, axis=1)



# CRP classification
def classify_crp(row):
    if pd.isna(row["LBXCRP"]):
        return np.nan
    crp = row["LBXCRP"]
    if crp < 1.0:
        return "Low risk"
    elif 1.0 <= crp < 3.0:
        return "Moderate risk"
    elif crp >= 3.0:
        return "High risk"
    else:
        return np.nan

df["CRP_Category"] = df.apply(classify_crp, axis=1)

# Create log_CRP for plotting (replace 0 with nan)
df["log_CRP"] = np.log(df["LBXCRP"].replace(0, np.nan))

# Drop SEQN column
df.drop(columns=["SEQN"], inplace=True)

# Restrict age to adults only
df = df[df["RIDAGEYR"] >= 18]

# Replace 'Unknown' with NaN in all columns
df.replace("Unknown", np.nan, inplace=True)

# Define categorical and numerical columns (for reference)
cat_cols = ["RIAGENDR", "Smoking", "Exercise", "BMI_category", "SES_Category", "Asthma_Status", "Spirometry_Category", "CRP_Category"]
num_cols = ["RIDAGEYR", "BMXBMI"]

print("\nSize of dataset: ", "Columns: ", len(df.columns), "Rows: ", len(df), "\n")
display(df[num_cols].describe())

print("\nMissing data")
print(df.isnull().sum())
print(df.columns.tolist())
```

```python
import pandas as pd

# -----------------------------
# Step 1: Define years and cycles
# -----------------------------
years = ["2007", "2009", "2011", "2013", "2015", "2017"]
cycles = ["E", "F", "G", "H", "I", "J"]

# -----------------------------
# Step 2: Load blood pressure data
# -----------------------------
bpx_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/BPX_{c}.xpt"
            for y, c in zip(years, cycles)]

df_bpx = pd.concat([pd.read_sas(url) for url in bpx_urls], ignore_index=True)
df_bpx = df_bpx[["SEQN", "BPXSY1", "BPXSY2", "BPXSY3", "BPXSY4"]]

# -----------------------------
# Step 3: Load asthma data
# -----------------------------
asthma_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/MCQ_{c}.xpt"
               for y, c in zip(years, cycles)]

df_asthma = pd.concat([pd.read_sas(url) for url in asthma_urls], ignore_index=True)
df_asthma = df_asthma[["SEQN", "MCQ010"]]  # MCQ010: "Ever told you had asthma?"

# -----------------------------
# Step 4: Merge datasets on SEQN
# -----------------------------
df_merged = pd.merge(df_bpx, df_asthma, on="SEQN", how="inner")

# Optional: Preview
print(df_merged.head())
```

```python
#Systolic blood pressure
import pandas as pd
import numpy as np

# Calculate average systolic blood pressure (ignoring NaNs)
df_merged['Mean_SBP'] = df_merged[['BPXSY1', 'BPXSY2', 'BPXSY3', 'BPXSY4']].mean(axis=1, skipna=True)

# Categorize based on clinical thresholds
def categorize_sbp(sbp):
    if pd.isna(sbp):
        return np.nan
    elif sbp < 120:
        return 'Normal'
    elif 120 <= sbp < 130:
        return 'Elevated'
    elif 130 <= sbp < 140:
        return 'Hypertension Stage 1'
    else:
        return 'Hypertension Stage 2'

df_merged['SBP_Category'] = df_merged['Mean_SBP'].apply(categorize_sbp)

# View distribution
print(df_merged['SBP_Category'].value_counts(dropna=False))
```

```python
# Define asthma classification function
def classify_asthma(row):
    if pd.isna(row["MCQ010"]):
        return np.nan
    elif row["MCQ010"] == 1:
        return "Asthma"
    elif row["MCQ010"] == 2:
        return "No asthma"
    elif row["MCQ010"] in [7, 9]:
        return np.nan
    return np.nan

# Apply classification to df_asthma
df_asthma["Asthma_Status"] = df_asthma.apply(classify_asthma, axis=1)

# Optionally drop MCQ010
df_asthma.drop(columns=["MCQ010"], inplace=True)

# View counts
print(df_asthma["Asthma_Status"].value_counts(dropna=False))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Map asthma status if not already done
df_merged['Asthma_Status'] = df_merged['MCQ010'].map({1: 'Asthma', 2: 'No asthma'})

# Step 2: Calculate mean SBP from BPXSY1–4
df_merged['Mean_SBP'] = df_merged[['BPXSY1', 'BPXSY2', 'BPXSY3', 'BPXSY4']].mean(axis=1, skipna=True)

# Step 3: Drop rows with missing values
df_sbp = df_merged.dropna(subset=['Mean_SBP', 'Asthma_Status'])

# Step 4: Group by asthma status and calculate mean SBP
mean_sbp_by_asthma = df_sbp.groupby('Asthma_Status')['Mean_SBP'].mean().reset_index()

# Step 5: Plot grouped bar chart
plt.figure(figsize=(7, 5))
sns.barplot(data=mean_sbp_by_asthma, x='Asthma_Status', y='Mean_SBP', palette='Set2')
plt.title('Mean Systolic Blood Pressure by Asthma Status')
plt.ylabel('Mean Systolic BP (mmHg)')
plt.xlabel('Asthma Status')
plt.tight_layout()
plt.show()
```

```python
#Two sample t-test to test if presence of asthma affects systolic blood pressure
from scipy.stats import ttest_ind

# Step 1: Clean the dataset
df_merged['Asthma_Status'] = df_merged['MCQ010'].map({1: 'Asthma', 2: 'No asthma'})
df_merged['Mean_SBP'] = df_merged[['BPXSY1', 'BPXSY2', 'BPXSY3', 'BPXSY4']].mean(axis=1, skipna=True)

# Step 2: Drop missing values
df_sbp = df_merged.dropna(subset=['Mean_SBP', 'Asthma_Status'])

# Step 3: Separate groups
asthma_group = df_sbp[df_sbp['Asthma_Status'] == 'Asthma']['Mean_SBP']
no_asthma_group = df_sbp[df_sbp['Asthma_Status'] == 'No asthma']['Mean_SBP']

# Step 4: Perform independent t-test
t_stat, p_val = ttest_ind(asthma_group, no_asthma_group, equal_var=False)

# Step 5: Report results
print("Independent t-test results:")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("✅ There is a statistically significant difference in SBP between asthmatics and non-asthmatics.")
else:
    print("❌ No statistically significant difference in SBP between asthmatics and non-asthmatics.")
```
