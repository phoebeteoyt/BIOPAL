# **4. Background and rationale**

## **4.1 Health issue or knowledge gap**

Chronic respiratory disease includes a range of conditions that affect multiple components of respiratory system, including the airways, parenchyma, and pulmonary vasculature. There are two types of chronic respiratory disease, communicable and noncommunicable. Communicable Chronic respiratory disease is long-term respiratory condition caused by infectious agents, while noncommunicable chronic respiratory disease are long-term respiratory conditions caused from lifestyle factors, chronic inflammation and environmental exposure. Some of the noncommunicable respiratory disease includes asthma, chronic obstructive pulmonary disease (COPD), lung cancer and others. Asthma is a major noncommunicable respiratory disease, and it has affected an estimated 262 million people in 2019, making it increasingly common these days, and it has caused 455,000 deaths according to WHO. It is also often under-diagnosed and under-treated in rural areas (with low income). To diagnose for the functioning of lungs especially when monitoring Chronic obstructive pulmonary disease, we can use spirometers to determine the ventilatory capacity of the lungs. Healthy lung size and function is dependent on four key variables, which are age, sex, height, and race, and the lung function values from the individual will be compared to normal patients with similar demographics, and if the value is below the normal range, it is considered abnormal. Some of the measurement used to interpret spirometry includes forced vital capacity (VC), forced expiratory volume in the first second (FEV 1), FEV1/FVC ratio and peak expiratory flow (PEF). In addition, elevated levels of C-reactive protein (CRP), a marker of low-grade systemic inflammation is also found associated in asthmatic individuals.

# **5. Research question and Objectives**

## **5.1 Research question**

- Using NHANES 2007-2018, among U.S. adults aged 18+, how do smoking status and BMI predict reduced FEV1/FVC ratio in individuals with and without prior asthma diagnosis?

## **5.2 Justification and novelty**

Reduced lung function leading to respiratory disease can be measured by the FEV1/FVC ratio and it is influenced by various lifestyle, environmental and demographic factors. Smoking is known for being a risk factor for respiratory disease, but different smoking status (e.g. occasional smoker, regular smoker, quite smoking less than or more than 1 year ago) has different impacts on the lungs. BMI has also been associated with impaired respiratory function and systemic inflammation, which can further affect respiratory health in both non-asthmatic and asthmatic individuals. There has been research on the independent effects of smoking and BMI on lung health, but not a lot have explored how these factors jointly influence FEV1/FVC ratio in subgroups defined by asthma status.

## **5.3 Contribution to existing literature**

Previous studies have shown that in non-obese individuals, BMI does not significantly affect lung function, whereas in obese individuals, BMI does affect lung function. It is also shown that non-asthmatic smokers face greater decline in FEV1 than non-asthmatic non-smokers, and asthmatic smokers face greater FEV1 decline than non-asthmatic smokers. Furthermore, it has been observed that increasing BMI is associated to lower FEV1/FVC ratios, suggesting that obesity is an independent factor that affects impairment of lung function. It however remains unclear if other risk factors such as smoking could influence this relationship, and whether asthmatics and non-asthmatics respond the same to changes in these independent factors. Through this research project, we are extending existing research by:

### **Importing data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Step 0: Define years and NHANES cycles
# -----------------------------
years = ["2007", "2009", "2011", "2013", "2015", "2017"]
cycles = ["E", "F", "G", "H", "I", "J"]

#Load gender, age, family income to poverty ratio data
demo_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/DEMO_{c}.xpt"
             for y, c in zip(years, cycles)]
df_demo = pd.concat([pd.read_sas(url) for url in demo_urls], ignore_index=True)

df_demo = df_demo[["SEQN", "RIAGENDR", "RIDAGEYR", "INDFMPIR"]]

#Load BMI data
bmi_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/BMX_{c}.xpt"
            for y, c in zip(years, cycles)]
df_bmi = pd.concat([pd.read_sas(url) for url in bmi_urls], ignore_index=True)

df_bmi = df_bmi[["SEQN", "BMXBMI"]]

#Load asthma status data
asthma_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/MCQ_{c}.xpt"
               for y, c in zip(years, cycles)]
df_asthma = pd.concat([pd.read_sas(url) for url in asthma_urls], ignore_index=True)

df_asthma = df_asthma[["SEQN", "MCQ010"]]  # MCQ010: "Ever told you had asthma?"

#Load Smoking status data
smq_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/SMQ_{c}.xpt"
            for y, c in zip(years, cycles)]
df_smq = pd.concat([pd.read_sas(url) for url in smq_urls], ignore_index=True)

df_smq = df_smq[["SEQN", "SMQ020", "SMQ040"]]  # Smoking status variables

# Load Spirometry (FEV1 & FVC) data
# -----------------------------
spirometry_years   = ["2007", "2009", "2011"]
spirometry_cycles  = ["E", "F", "G"]          # Only 2007–2012 have spirometry
spx_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/SPX_{c}.xpt"
            for y, c in zip(spirometry_years, spirometry_cycles)]

df_spiro = pd.concat([pd.read_sas(url) for url in spx_urls])

# keep baseline FEV1 & FVC
df_spiro = df_spiro[["SEQN", "SPXNFEV1", "SPXNFVC"]]
df_spiro = df_spiro.rename(columns={"SPXNFEV1": "FEV1", "SPXNFVC": "FVC"})

paq_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/PAQ_{c}.xpt"
            for y, c in zip(years, cycles)]
df_paq = pd.concat([pd.read_sas(url) for url in paq_urls], ignore_index=True)

df_paq = df_paq[["SEQN", "PAD660", "PAD675", "PAD680"]]

# Load CRP / hs-CRP data
import pandas as pd
import requests
import io
def crp_url_and_var(year: str, cycle_letter: str):
    if cycle_letter in {"E", "F"}:          # 2007–2010 cycles
        return (
            f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/CRP_{cycle_letter}.xpt",
            "LBXCRP",
        )
    elif cycle_letter in {"I", "J"}:        # 2015–2018 cycles
        return (
            f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/HSCRP_{cycle_letter}.xpt",
            "LBXHSCRP",
        )
    else:                                   # No CRP measured
        return None

# ------------------------------------------------------------
# Download and combine CRP / hs-CRP for all requested cycles
# ------------------------------------------------------------
crp_frames = []
for year, letter in zip(years, cycles):
    info = crp_url_and_var(year, letter)
    if info is None:
        continue                                   # skip cycles without CRP

    url, varname = info
    try:
        buffer = io.BytesIO(requests.get(url, timeout=60).content)
        tmp = pd.read_sas(buffer, format="xport")[["SEQN", varname]]
        tmp = tmp.rename(columns={varname: "LBXCRP"})
        crp_frames.append(tmp)
    except Exception:
        # silently ignore files that fail to load
        pass

# Concatenate downloaded frames; create an empty placeholder if none loaded
df_crp = (
    pd.concat(crp_frames, ignore_index=True)
    if crp_frames
    else pd.DataFrame(columns=["SEQN", "LBXCRP"])
)


#Load blood pressure data
# -----------------------------
bpx_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/BPX_{c}.xpt"
            for y, c in zip(years, cycles)]

df_bpx = pd.concat([pd.read_sas(url) for url in bpx_urls], ignore_index=True)
df_bpx = df_bpx[["SEQN", "BPXSY1", "BPXSY2", "BPXSY3", "BPXSY4"]]
```

### **Merging data**

```python
# ---------------------------------------------
# Merge all NHANES tables on SEQN (left joins)
# ---------------------------------------------
df_merged = (
    df_demo
      .merge(df_bmi,    on="SEQN", how="left")
      .merge(df_asthma, on="SEQN", how="left")
      .merge(df_smq,    on="SEQN", how="left")
      .merge(df_spiro,  on="SEQN", how="left")
      .merge(df_paq,    on="SEQN", how="left")
      .merge(df_crp,    on="SEQN", how="left")
      .merge(df_bpx,    on="SEQN", how="left")
)

# Optional: derived variable
df_merged["FEV1_FVC_ratio"] = df_merged["FEV1"] / df_merged["FVC"]

# Quick check
print(df_merged.shape)
df_merged.head()
```

### **Categorizing data**

```python
import pandas as pd
import numpy as np
from pandas.api.types import CategoricalDtype

# ------------------------------------------------------------
# Helper: exercise cleaner + full categoriser
# ------------------------------------------------------------
def add_categories(df):
    out = df.copy()

    # ---------- Demographics ----------
    out["Gender"] = out["RIAGENDR"].map({1: "Male", 2: "Female"}).astype("category")
    out["AgeGroup"] = pd.cut(
        out["RIDAGEYR"],
        bins=[17, 29, 39, 49, 59, 69, 79, 89, np.inf],
        labels=["18-29", "30-39", "40-49", "50-59",
                "60-69", "70-79", "80-89", "90+"]
    ).astype("category")
    out["SES_Category"] = out["INDFMPIR"].apply(
        lambda x: "<1" if x < 1 else
                  "1–<2" if x < 2 else
                  "2–<5" if x < 5 else "≥5"
    ).astype("category")
    out["BMI_Category"] = pd.cut(
        out["BMXBMI"], [0, 18.5, 25, 30, np.inf],
        labels=["Underweight", "Normal", "Overweight", "Obese"]
    ).astype("category")

    # ---------- Smoking ----------
    def classify_smoking(row):
        if pd.isna(row["SMQ020"]):
            return np.nan
        if row["SMQ020"] == 2:
            return "Never smoked"
        if row["SMQ020"] == 1:
            if pd.isna(row["SMQ040"]):
                return np.nan
            if row["SMQ040"] == 1:
                return "Regular smoker"
            if row["SMQ040"] == 2:
                return "Occasional smoker"
            if row["SMQ040"] == 3:
                return "Reformed non-smoker"
        return np.nan
    out["Smoking"] = out.apply(classify_smoking, axis=1).astype("category")

    # ---------- Asthma ----------
    out["Asthma_Status"] = out["MCQ010"].map(
        {1: "Asthma", 2: "No Asthma"}
    ).astype("category")

    # ---------- Exercise (clean PAD* codes then classify) ----------
    # Replace NHANES skip / missing codes with NaN
    repl_long = {777: np.nan, 999: np.nan}
    repl_short = {77: np.nan, 99: np.nan}
    out["PAD660"] = out["PAD660"].replace(repl_long)
    out["PAD675"] = out["PAD675"].replace(repl_long)
    out["PAD680"] = out["PAD680"].replace(repl_short)

    def classify_activity(row):
        vig = not pd.isna(row["PAD660"]) and row["PAD660"] > 0
        mod = not pd.isna(row["PAD675"]) and row["PAD675"] > 0
        if vig and mod:
            return "Moderate & vigorous"
        elif vig:
            return "Vigorous only"
        elif mod:
            return "Moderate only"
        else:
            return "None"
    out["Exercise"] = out.apply(classify_activity, axis=1).astype("category")

    # ---------- Biomarkers ----------
    out["CRP_cat"] = pd.cut(
        out["LBXCRP"], [-np.inf, 1, 3, np.inf],
        labels=["Low risk (<1)", "Moderate risk (1-3)", "High risk (>3)"]
    ).astype("category")

    out["SBP_mean"] = out[["BPXSY1", "BPXSY2", "BPXSY3", "BPXSY4"]].mean(axis=1)
    out["SBP_cat"] = pd.cut(
        out["SBP_mean"], [0, 120, 130, 140, 180, np.inf],
        labels=["Normal (<120)", "Elevated (120-129)",
                "Stage 1 HTN", "Stage 2 HTN", "Crisis (≥180)"]
    ).astype("category")

    # ---------- Spirometry ----------
    out["FEV1_FVC_ratio"] = out["FEV1"] / out["FVC"]
    out["FEV1_FVC_ratio"] = pd.Series(
        np.where(out["FEV1_FVC_ratio"] < 0.70, "Obstructive", "Normal"),
        index=out.index
    ).astype("category")

    return out

# ------------------------------------------------------------
# Apply categorisation
# ------------------------------------------------------------
df_cat = add_categories(df_merged)

# ------------------------------------------------------------
# Print categories + counts
# ------------------------------------------------------------
cat_cols = [
    "Gender", "AgeGroup", "SES_Category", "BMI_Category",
    "Smoking", "Asthma_Status", "Exercise",
    "CRP_cat", "SBP_cat", "FEV1_FVC_ratio"
]

for col in cat_cols:
    print(f"\n===== {col} =====")
    if isinstance(df_cat[col].dtype, CategoricalDtype):
        print("Categories:", list(df_cat[col].cat.categories))
    else:
        print("Categories: (not categorical)")
    print("Counts:")
    print(df_cat[col].value_counts(dropna=False))
```

### Using Kruskal Wallis test followed by Dunett Post-hoc test,

```python

from scipy.stats import shapiro

# Prepare FEV1/FVC as continuous ratio
df_test = df_clean.dropna(subset=["Smoking", "FEV1", "FVC"]).copy()
df_test["FEV1_FVC_cont"] = df_test["FEV1"] / df_test["FVC"]

# Shapiro–Wilk test for each smoking group
print("Shapiro–Wilk Normality Test for FEV1/FVC Ratio by Smoking Group\n")
for group in df_test["Smoking"].dropna().unique():
    values = df_test[df_test["Smoking"] == group]["FEV1_FVC_cont"]
    if len(values) > 5000:
        values = values.sample(5000, random_state=42)  # Limit for Shapiro-Wilk
    stat, p = shapiro(values)
    print(f"{group}: W = {stat:.4f}, p = {p:.4f} (n = {len(values)})")
```

We can see that the W values of the different smoking categories are close to 1 which means that data is close to normal distribution. However, the p values is less than 0.05, so we reject the null hypothesis, meaning that the data is significantly different from a normal distribution. Thus, we will carry out a Kruskal Wallis test

```python
from scipy.stats import kruskal

# Group FEV1/FVC by smoking group
groups = [g["FEV1_FVC_cont"].values for name, g in df_test.groupby("Smoking")]

# Perform Kruskal–Wallis test
stat, p = kruskal(*groups)
print(f"\nKruskal–Wallis Test:\nH = {stat:.3f}, p = {p:.4f}")
```

There is a statistically significant difference in the FEV1/FVC ratio among the different smoking groups.

## **Association of SES with presence of Asthma**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Step 1: Income category binning
bins = [0, 1, 2, 3, 5, 10, 100]
labels = ['<1 (Below poverty)', '1–<2', '2–<3', '3–<5', '5–<10', '10+']
df['Income_Category'] = pd.cut(df['INDFMPIR'], bins=bins, labels=labels)

# Step 2: Map asthma status
df['Asthma_Status'] = df['MCQ010'].map({1: "Yes", 2: "No"}).astype("category")

# Step 3: Clean and display contingency table
df_ct = df[["Income_Category", "Asthma_Status"]].dropna()
contingency = pd.crosstab(df_ct["Income_Category"], df_ct["Asthma_Status"])

print("Contingency table:\n", contingency)

# Step 4: Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency)
print(f"\nChi-square statistic = {chi2:.2f},  p-value = {p:.4f}")
if p < 0.05:
    print("Result: Significant association.")
else:
    print("Result: No significant association.")

# Step 5: Stacked bar chart
contingency.plot(kind='bar', stacked=True, figsize=(10,6), color=["skyblue", "salmon"])
plt.title("Asthma Presence by Income Category")
plt.xlabel("Income-to-Poverty Ratio")
plt.ylabel("Number of Participants")
plt.legend(title="Asthma Status")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## **Mean CRP affected by Smoking status across different age groups**
###Using Multiple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# 1.  Work from the cleaned / categorised table
# ------------------------------------------------------------
df_plot = df_clean.copy()            # or df_cat / df_merged

# ------------------------------------------------------------
# 2.  Ensure Smoking column is present
# ------------------------------------------------------------
if "Smoking" not in df_plot.columns:
    raise KeyError("Column 'Smoking' not found – run add_categories() first.")

# ------------------------------------------------------------
# 3.  Build age-group bands (18-29, 30-44, 45-65)
# ------------------------------------------------------------
df_plot = df_plot[df_plot["RIDAGEYR"].between(18, 65)].copy()

age_bins   = [18, 29, 44, 65]
age_labels = ["18–29", "30–44", "45–65"]
df_plot["Age_Group"] = pd.cut(
    df_plot["RIDAGEYR"],
    bins=age_bins,
    labels=age_labels,
    right=True,
    include_lowest=True
)

# ------------------------------------------------------------
# 4.  Keep complete rows: CRP + Smoking + Age_Group
# ------------------------------------------------------------
df_plot = df_plot.dropna(subset=["LBXCRP", "Smoking", "Age_Group"])

# ------------------------------------------------------------
# 5.  Compute summary stats (mean ± 95 % CI)
# ------------------------------------------------------------
summary = (
    df_plot.groupby(["Age_Group", "Smoking"])["LBXCRP"]
           .agg(mean="mean", count="count", std="std")
           .reset_index()
)

summary["sem"]      = summary["std"] / np.sqrt(summary["count"])
summary["ci95_lo"]  = summary["mean"] - 1.96 * summary["sem"]
summary["ci95_hi"]  = summary["mean"] + 1.96 * summary["sem"]

# enforce categorical order
age_dtype = pd.CategoricalDtype(categories=age_labels, ordered=True)
smoking_order = ["Never smoked", "Reformed non-smoker",
                 "Occasional smoker", "Regular smoker"]
smoke_dtype = pd.CategoricalDtype(categories=smoking_order, ordered=True)

summary["Age_Group"] = summary["Age_Group"].astype(age_dtype)
summary["Smoking"]   = summary["Smoking"].astype(smoke_dtype)
summary = summary.sort_values(["Age_Group", "Smoking"])

# ------------------------------------------------------------
# 6.  Bar-plot with 95 % CI
# ------------------------------------------------------------
fig, ax = plt.subplots(figsize=(10, 6))

bar_w      = 0.18
x_pos      = np.arange(len(age_labels))

for i, status in enumerate(smoking_order):
    sub = summary[summary["Smoking"] == status]
    ax.bar(
        x_pos + i * bar_w,
        sub["mean"],
        width=bar_w,
        yerr=np.vstack([
            sub["mean"] - sub["ci95_lo"],
            sub["ci95_hi"] - sub["mean"]
        ]),
        capsize=4,
        label=status
    )

# Axis / legend
ax.set_xticks(x_pos + bar_w * (len(smoking_order) - 1) / 2)
ax.set_xticklabels(age_labels)
ax.set_xlabel("Age Group")
ax.set_ylabel("Mean CRP (mg/L)")
ax.set_title("Mean CRP by Smoking Status Across Age Groups (18–65 yr)")
ax.legend(title="Smoking Status", fontsize=9)
ax.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

In this bar chart we observe the general trend as age increases, individuals' mean CRP levels increase.

When we look closely at the oldest age group of 45-65, occasional smokers and regular smokers have higher mean CRP values than non-smokers and reformed non-smokers. This is not seen in the younger age groups. This suggests that the inflammatory effects from smoking could intensify as you age.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.api.types import CategoricalDtype
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm

# ------------------------------------------------------------
# 1.  Start from cleaned / categorised table
# ------------------------------------------------------------
df_ = df_clean.copy()                       # or df_cat / df_merged

# ------------------------------------------------------------
# 2.  Ensure Exercise, Asthma_Status, log_CRP exist
# ------------------------------------------------------------
if "Asthma_Status" not in df_.columns:
    df_["Asthma_Status"] = df_["MCQ010"].map({1: "Yes", 2: "No"})

if "Exercise" not in df_.columns:
    raise KeyError("Column 'Exercise' not found – run add_categories() first.")

df_["log_CRP"] = np.log(df_["LBXCRP"].clip(lower=1e-4))

# 4-level activity order
ex_order = ["None", "Moderate only", "Vigorous only", "Moderate & vigorous"]
ex_dtype = CategoricalDtype(categories=ex_order, ordered=True)
df_["Exercise"] = df_["Exercise"].astype(ex_dtype)

# ------------------------------------------------------------
# 3.  Keep complete rows
# ------------------------------------------------------------
df_pa = df_.dropna(subset=["log_CRP", "Exercise", "Asthma_Status"])

# ------------------------------------------------------------
# 4.  Box-plot (log-CRP × Exercise × Asthma)
# ------------------------------------------------------------
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df_pa,
    x="Exercise",
    y="log_CRP",
    hue="Asthma_Status",
    palette="Set1"
)
plt.title("log(CRP) by Physical Activity and Asthma Status")
plt.xlabel("Physical-activity category")
plt.ylabel("log(CRP mg/L)")
plt.xticks(rotation=20)
plt.legend(title="Asthma Status")
plt.tight_layout()
plt.show()
```

In general, more physical activity is associated to lower logCRP values, and non-asthmatics tend to have lower logCRP values than asthmatics.

To verify this relationship we conducted a two-way ANOVA.

We found that both physical activity and asthma status significantly influenced log(CRP) independently. However, the interaction between physical activity and asthma status was not significant.

```python
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Step 1: Prepare data
df_reg = df_clean.copy()
df_reg["log_CRP"] = np.log(df_reg["LBXCRP"].clip(lower=1e-4))

# Step 2: Drop missing values
df_reg = df_reg.dropna(subset=["log_CRP", "Exercise", "Asthma_Status"])

# Step 3: Fit the MLR model with interaction
model = ols("log_CRP ~ C(Exercise) * C(Asthma_Status)", data=df_reg).fit()

# Step 4: Output the summary
print(model.summary())
# ✅ Interpretation for MLR: Exercise × Asthma_Status on log(CRP)
p_exercise  = model.pvalues.get("C(Exercise)", np.nan)  # may be NaN if multiple levels
p_asthma    = model.pvalues.get("C(Asthma_Status)", np.nan)
p_interact  = model.pvalues.filter(like="C(Exercise):C(Asthma_Status)").min()

# Main effect: Exercise
if not np.isnan(p_exercise):
    if p_exercise < 0.05:
        print("\n✅ There is a **statistically significant main effect** of Exercise on log(CRP).")
    else:
        print("\n❌ There is **no significant main effect** of Exercise on log(CRP).")

# Main effect: Asthma_Status
if p_asthma < 0.05:
    print("✅ There is a **statistically significant main effect** of Asthma_Status on log(CRP).")
else:
    print("❌ There is **no significant main effect** of Asthma_Status on log(CRP).")

# Interaction
if p_interact < 0.05:
    print("✅ There is a **statistically significant interaction** between Exercise and Asthma_Status on log(CRP).")
else:
    print("❌ There is **no significant interaction** between Exercise and Asthma_Status on log(CRP).")

    # Interpretation for Exercise main effect (overall factor)
if p_exercise < 0.05:
    print("✅ There is a **statistically significant main effect** of Exercise on log(CRP).")
else:
    print("❌ There is **no significant main effect** of Exercise on log(CRP).")

# Specific contrasts (only if you want to mention notable ones)
print("\n📊 Pairwise contrasts vs. reference group:")
print("   - Moderate only: ❌ No significant difference (p = 0.151)")
print("   - None: ✅ Significantly higher CRP (~57% higher, p < 0.001)")
print("   - Vigorous only: ❌ No significant difference (p = 0.824)")
```

### R² = 0.012, Adjusted R² = 0.012 → Only about 1.2% of the variance in log(CRP) is explained by Exercise, Asthma_Status, and their interaction.

### The significance F value is lower than the significance level of 0.05 → The overall model is still statistically significant, even though the effect size is small.

### At the factor level, physical activity does not reliably predict CRP in this model.

### However when we look at looking at the specific contrasts versus the reference activity group, we can tell that while the factor ‘Exercise’ isn’t significant overall, participants reporting no exercise show meaningfully higher CRP, consistent with the idea that inactivity relates to low-grade inflammation. But since the overall test wasn’t significant, we should just take this observation lightly.

## **Effect of Gender on mean CRP**
###Using Multiple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# 1. Start from cleaned dataframe
# ------------------------------------------------------------
df_plot = df_clean.copy()

# ------------------------------------------------------------
# 2. Map gender (if not present)
# ------------------------------------------------------------
if "Gender" not in df_plot.columns:
    df_plot["Gender"] = df_plot["RIAGENDR"].map({1: "Male", 2: "Female"})

# ------------------------------------------------------------
# 3. Keep rows with valid CRP + gender
# ------------------------------------------------------------
df_plot = df_plot.dropna(subset=["LBXCRP", "Gender"])

# ------------------------------------------------------------
# 4. Summary stats (mean ± 95 % CI)
# ------------------------------------------------------------
summary = (
    df_plot.groupby("Gender")["LBXCRP"]
           .agg(mean="mean", count="count", std="std")
           .reset_index()
)
summary["sem"]     = summary["std"] / np.sqrt(summary["count"])
summary["ci95_lo"] = summary["mean"] - 1.96 * summary["sem"]
summary["ci95_hi"] = summary["mean"] + 1.96 * summary["sem"]

# Error-bar distances
yerr = np.vstack([
    summary["mean"] - summary["ci95_lo"],
    summary["ci95_hi"] - summary["mean"]
])

# ------------------------------------------------------------
# 5. Plot
# ------------------------------------------------------------
plt.figure(figsize=(6, 5))
plt.bar(
    summary["Gender"],
    summary["mean"],
    yerr=yerr,
    capsize=8,
    color=["skyblue", "lightpink"],
    edgecolor="black"
)
plt.title("Mean CRP by Gender (95 % CI)")
plt.xlabel("Gender")
plt.ylabel("CRP (mg/L)")
# Uncomment next line if CRP is highly skewed
# plt.yscale("log")
plt.tight_layout()
plt.show()
```

We can observe that mean CRP in females are significantly higher than that in males. This suggests that gender is associated to systemic inflammation.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# --- Multiple Linear Regression: CRP ~ Gender + confounders ---
# Log-transform CRP to reduce skew
df_reg = df_clean.copy()
df_reg["log_CRP"] = np.log(df_reg["LBXCRP"].clip(lower=1e-4))

# Fit model (adjust for BMI, Age, Smoking, Asthma)
model = ols("log_CRP ~ C(Gender) + BMXBMI + RIDAGEYR + C(Smoking) + C(Asthma_Status)", data=df_reg).fit()

print(model.summary())

# --- Interpretation ---
p_val_gender = model.pvalues["C(Gender)[T.Male]"]

if p_val_gender < 0.05:
    print(f"\n✅ There is a **statistically significant difference** in log(CRP) between genders "
          f"(p = {p_val_gender:.4f}). Adjusting for BMI, age, smoking, and asthma, "
          f"the coefficient for male gender is {model.params['C(Gender)[T.Male]']:.4f}, "
          "meaning males tend to have lower CRP values on average compared to females "
          "(in log-scale).")
else:
    print(f"\n❌ There is **no significant difference** in log(CRP) between genders "
          f"(p = {p_val_gender:.4f}) after adjusting for BMI, age, smoking, and asthma.")

```

## **Effect of Asthma presence on Systolic blood pressure**
### Using Mann-Whitney test

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.api.types import CategoricalDtype

# ------------------------------------------------------------
# 1.  Work from your cleaned / categorised dataframe
# ------------------------------------------------------------
df_bp = df_clean.copy()                 # or df_cat / df_merged

# ------------------------------------------------------------
# 2.  Ensure Asthma_Status label exists
# ------------------------------------------------------------
df_bp["Asthma_Status"] = df_bp["MCQ010"].map({1: "Asthma", 2: "No asthma"})

# enforce categorical order
asthma_dtype = CategoricalDtype(categories=["Asthma", "No asthma"], ordered=True)
df_bp["Asthma_Status"] = df_bp["Asthma_Status"].astype(asthma_dtype)

# ------------------------------------------------------------
# 3.  Compute mean SBP across up to four readings
# ------------------------------------------------------------
df_bp["SBP_mean"] = df_bp[["BPXSY1", "BPXSY2", "BPXSY3", "BPXSY4"]].mean(axis=1, skipna=True)

# ------------------------------------------------------------
# 4.  Keep complete cases (Mean_SBP + asthma label)
# ------------------------------------------------------------
df_bp = df_bp.dropna(subset=["SBP_mean", "Asthma_Status"])

# ------------------------------------------------------------
# 5.  Summary stats (mean ± 95 % CI)
# ------------------------------------------------------------
summary = (
    df_bp.groupby("Asthma_Status")["SBP_mean"]
         .agg(mean="mean", count="count", std="std")
         .reset_index()
)
summary["sem"]      = summary["std"] / np.sqrt(summary["count"])
summary["ci95_lo"]  = summary["mean"] - 1.96 * summary["sem"]
summary["ci95_hi"]  = summary["mean"] + 1.96 * summary["sem"]

# ------------------------------------------------------------
# 6.  Bar-chart with 95 % CI error bars
# ------------------------------------------------------------
plt.figure(figsize=(7, 5))
sns.barplot(
    data=df_bp,
    x="Asthma_Status",
    y="SBP_mean",
    ci=95,
    capsize=0.15,
    palette="Set2",
    edgecolor="black"
)
plt.title("Mean Systolic Blood Pressure by Asthma Status")
plt.xlabel("Asthma Status")
plt.ylabel("Mean SBP (mmHg)")
plt.tight_layout()
plt.show()

#Two sample t-test to test if presence of asthma affects systolic blood pressure
from scipy.stats import ttest_ind

# Step 1: Clean the dataset
df_merged['Asthma_Status'] = df_merged['MCQ010'].map({1: 'Asthma', 2: 'No asthma'})
df_merged['Mean_SBP'] = df_merged[['BPXSY1', 'BPXSY2', 'BPXSY3', 'BPXSY4']].mean(axis=1, skipna=True)

# Step 2: Drop missing values
df_sbp = df_merged.dropna(subset=['Mean_SBP', 'Asthma_Status'])

# Step 3: Separate groups
asthma_group = df_sbp[df_sbp['Asthma_Status'] == 'Asthma']['Mean_SBP']
no_asthma_group = df_sbp[df_sbp['Asthma_Status'] == 'No asthma']['Mean_SBP']

# Step 4: Perform independent t-test
t_stat, p_val = ttest_ind(asthma_group, no_asthma_group, equal_var=False)

# Step 5: Report results
print("Independent t-test results:")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("✅ There is a statistically significant difference in SBP between asthmatics and non-asthmatics.")
else:
    print("❌ No statistically significant difference in SBP between asthmatics and non-asthmatics.")

```

In non-asthmatics, the mean systolic blood pressure is higher than that of asthmatics. In the graph, the difference does not look significant. A t-test has been conducted further, and the results have shown that there is indeed a significant difference in systolic blood pressure between asthmatics and non-asthmatics.

```python
from scipy.stats import mannwhitneyu

# Ensure clean data
df_test = df_clean.dropna(subset=["Asthma_Status", "SBP_mean"]).copy()

# Split groups
asthma_vals = df_test[df_test["Asthma_Status"] == "Asthma"]["SBP_mean"]
no_asthma_vals = df_test[df_test["Asthma_Status"] == "No Asthma"]["SBP_mean"]

# Mann–Whitney U test
stat, p = mannwhitneyu(asthma_vals, no_asthma_vals, alternative="two-sided")

# Print results
print("Mann–Whitney U Test: SBP by Asthma Status\n")
print(f"Statistic = {stat:.3f}")
print(f"p-value = {p:.4f}")
print(f"Mean SBP (Asthma): {asthma_vals.mean():.2f}")
print(f"Mean SBP (No Asthma): {no_asthma_vals.mean():.2f}")

# Interpretation
if p < 0.05:
    print("\n✅ There is a **significant difference** in systolic blood pressure between individuals with and without asthma.")
else:
    print("\n❌ There is **no significant difference** in systolic blood pressure between individuals with and without asthma.")
```

## **8.2 Future Work**
### **Tests for further analysis:**
**1. Survey-weighted logistic regression**

This method would translate our continuous measures of lung function (FEV1/FVC) and inflammation (CRP) into clinically relevant categories such as “low lung function” or “high inflammation.” It would allow us to estimate the odds of being in these high-risk groups based on smoking, BMI, and asthma status, making our results more applicable for risk prediction and screening purposes.


**2. Mediation analysis**

This would explore whether inflammation (CRP) acts as a mediator in the relationship between BMI and systolic blood pressure, helping to explain the biological pathway that links weight to cardiovascular strain.

**3. Multiple regression**

By including CRP, lung function, BMI, smoking, and asthma together in a single model for systolic blood pressure, we could determine which factors have the strongest overall effect when controlling for the others.

**4. Temporal trend analysis**

This would examine whether the strength or direction of the associations between key risk factors and our three indicators (FEV1/FVC, CRP, SBP) has changed across different NHANES survey cycles, helping identify evolving public health patterns.
